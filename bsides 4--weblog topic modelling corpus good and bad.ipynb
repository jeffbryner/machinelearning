```python
import sys
sys.path.reverse()
from gensim import corpora, models, similarities
```

```python
#learn words
dictionary = corpora.Dictionary(line.decode('ascii','ignore').lower().split() for line in open('www.mozilla.org.access_2013-12-31-13.good').readlines())
```

```python
dictionary.token2id
```

```python
len(dictionary)
```

```python
#learn some bad words from https://code.google.com/p/fuzzdb/source/browse/trunk/attack-payloads/all-attacks/all-attacks-unix.txt
dictionary.add_documents(line.lower().split() for line in open('web-attacks-unix.txt').readlines())
```

```python
dictionary.token2id
```

```python
len(dictionary)
```

```python
#make a corpus of good words only
corpusGood=[dictionary.doc2bow(line.decode('ascii','ignore').lower().split()) for line in open('www.mozilla.org.access_2013-12-31-13.good').readlines()[0:20000]]
```

```python
len(corpusGood)
```

```python
#latent semantic index model http://en.wikipedia.org/wiki/Latent_semantic_indexing
lsiGood = models.LsiModel(corpusGood, id2word=dictionary,num_topics=1000)
lsiGood.print_topics(1000)[5]
```

```python
agoodhit=r'''/thunderbird/js/jquery/jquery-1.5.1.min.js'''
```

```python
vecGood = dictionary.doc2bow(agoodhit.lower().split())
veclsiGood = lsiGood[vecGood] # convert the query to LSI space
```

```python
len(veclsiGood)
```

```python
from operator import itemgetter
sorted(veclsiGood, key=itemgetter(1),reverse=True)
```

```python
lsiGood.print_topics(1000)[4]
```

```python
#create a bad corpus
corpusBad=[dictionary.doc2bow(line.lower().split()) for line in open('web-attacks-unix.txt').readlines() ]
```

```python
lsiBad = models.LsiModel(corpusBad,id2word=dictionary,num_topics=1000)
vecBad = dictionary.doc2bow(agoodhit.lower().split())
veclsiBad = lsiBad[vecBad]
lsiBad.print_topics()
```

```python
abadhit=r'''GET ../../../../../../../../../../../../etc/passwd'''
```

```python
vecBad = dictionary.doc2bow(abadhit.lower().split())
veclsiBad = lsiBad[vecBad]
sorted(veclsiBad, key=itemgetter(1),reverse=True)
```

```python
lsiBad.print_topics(1000)[117]
```

```python
#if you didn't know..is the hit good or bad?
ahit=r'''POST ../../../../../../../../../../../../etc/passwd'''
vecHit = dictionary.doc2bow(ahit.lower().split())
veclsiBad = lsiBad[vecHit]
veclsiGood = lsiGood[vecHit]
#compare matches.
badMatches=sorted(veclsiBad, key=itemgetter(1),reverse=True)
goodMatches=sorted(veclsiGood, key=itemgetter(1),reverse=True)
```

```python
print(len(goodMatches))
print(len(badMatches))
```

```python
print(badMatches[0],goodMatches[0])
```
